# Dataset Overview

```python
# Import all the neccessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("seaborn-v0_8")
```

```python
# Load the weather dataset
# This dataset derived from the National Oceanic and Atmospheric Administration (NOAA) daily weather observations contains daily meteorological observations from the weather station at:
# Station ID: USW00094789, Location: JFK International Airport New York, Period: Jan 2010– Dec 2025
df = pd.read_csv("weather_data.csv")
# View first few rows of the dataset
df.head()
```

```python
# Display Dataset Shape
print("\nDataset Shape (Rows, Columns):", df.shape)

# Display Dataset Information
print("\nDataset Information:")
df.info()

# Display column names
print("\nColumn Names:")
print(df.columns)

# Display data-types for each column
print("\nData Types:")
print(df.dtypes)
```

```python
# Statistical Summary
df.describe()
```

```python
# Check how many missing values exist in each column
print("\nMissing Values in Each Column:")
print(df.isna().sum())
```

# Datetime Conversion

```python
# Convert DATE column into datetime format so we can analyze the dataset as a time series

df["DATE"] = pd.to_datetime(df["DATE"], dayfirst=True)

# Sort dataset by time just to ensure chronological order
df = df.sort_values("DATE")

# Create some helper columns that help us explore seasonality
df["YEAR"] = df["DATE"].dt.year
df["MONTH"] = df["DATE"].dt.month
df["DAY_OF_YEAR"] = df["DATE"].dt.dayofyear
```

# Time Series Visualization

```python
# The plots below illustrate how maximum and minimum temperatures change over time

# Maximum temperature over time
plt.figure()
plt.plot(df["DATE"], df["TMAX"])
plt.title("Maximum Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("TMAX")
plt.show()

# Minimum temperature over time
plt.figure()
plt.plot(df["DATE"], df["TMIN"])
plt.title("Minimum Temperature Over Time")
plt.xlabel("Date")
plt.ylabel("TMIN")
plt.show()
```

# Temperature Distribution Analysis

```python
# Distribution helps us understand if data is skewed, normal or contains extreme values.

plt.figure()
plt.hist(df["TMAX"], bins=40)
plt.title("Distribution of Maximum Temperature")
plt.xlabel("TMAX")
plt.ylabel("Frequency")
plt.show()

plt.figure()
plt.hist(df["TMIN"], bins=40)
plt.title("Distribution of Minimum Temperature")
plt.xlabel("TMIN")
plt.ylabel("Frequency")
plt.show()
```

# Temperature Variable Relationship

```python
# TMIN and TMAX relationship
plt.figure()
plt.scatter(df["TMIN"], df["TMAX"])
plt.title("Relationship Between TMIN and TMAX")
plt.xlabel("TMIN")
plt.ylabel("TMAX")
plt.show()
```

# Weather Variables Vs Temperature

```python
# Precipitation vs Temperature
plt.figure()
plt.scatter(df["PRCP"], df["TMAX"])
plt.title("Precipitation vs Maximum Temperature")
plt.xlabel("PRCP")
plt.ylabel("TMAX")
plt.show()


# Wind speed vs Temperature
plt.figure()
plt.scatter(df["AWND"], df["TMAX"])
plt.title("Average Wind Speed vs TMAX")
plt.xlabel("AWND")
plt.ylabel("TMAX")
plt.show()


# Snow vs minimum temperature
plt.figure()
plt.scatter(df["SNOW"], df["TMIN"])
plt.title("Snowfall vs Minimum Temperature")
plt.xlabel("SNOW")
plt.ylabel("TMIN")
plt.show()
```

# Annual Temperature Trends

```python
# This helps identify long-term temperature patterns

yearly_temp = df.groupby("YEAR")[["TMAX", "TMIN"]].mean()

plt.figure()
plt.plot(yearly_temp.index, yearly_temp["TMAX"])
plt.title("Average Yearly Maximum Temperature")
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.show()

plt.figure()
plt.plot(yearly_temp.index, yearly_temp["TMIN"])
plt.title("Average Yearly Minimum Temperature")
plt.xlabel("Year")
plt.ylabel("Temperature")
plt.show()
```

# Monthly Seasonality Analysis

```python
# Weather data is highly seasonal
# This plot shows average temperature across months

monthly_temp = df.groupby("MONTH")[["TMAX", "TMIN"]].mean()

plt.figure()
plt.plot(monthly_temp.index, monthly_temp["TMAX"])
plt.title("Average Monthly TMAX (Seasonality)")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.show()

plt.figure()
plt.plot(monthly_temp.index, monthly_temp["TMIN"])
plt.title("Average Monthly TMIN (Seasonality)")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.show()
```

# Rolling Temperature Analysis

```python
# Rolling averages help visualize general trends without daily noise

df["TMAX_30DAY_ROLL"] = df["TMAX"].rolling(window=30).mean()

plt.figure()
plt.plot(df["DATE"], df["TMAX"], alpha=0.3)
plt.plot(df["DATE"], df["TMAX_30DAY_ROLL"])
plt.title("TMAX with 30-Day Rolling Average")
plt.xlabel("Date")
plt.ylabel("Temperature")
plt.show()
```

# Correlation Analysis

```python
# Correlation heatmap
# Selecting only the most relevant numerical variables
# Select important numerical weather variables
corr_features = [
    "TMAX",
    "TMIN",
    "TAVG",
    "PRCP",
    "SNOW",
    "SNWD",
    "AWND",
    "WDF2",
    "WDF5",
    "WSF2",
    "WSF5"
]

# Compute correlation matrix
corr_matrix = df[corr_features].corr()

# Plot heatmap
plt.figure(figsize=(8,6))

sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)

plt.title("Correlation Matrix of Key Weather Variables")

plt.show()
```

# Lag Feature Exploration

```python
# Create lag just for exploration this helps understand how past temperature influences future

df["TMAX_LAG1"] = df["TMAX"].shift(1)

plt.figure()
plt.scatter(df["TMAX_LAG1"], df["TMAX"])
plt.title("Lag Relationship (Yesterday vs Today TMAX)")
plt.xlabel("Yesterday TMAX")
plt.ylabel("Today TMAX")
plt.show()
```
